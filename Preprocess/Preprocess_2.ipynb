```python
df_crop['Crop'] = df_crop['Crop'].str.lower()
df['label'] = df['label'].str.lower()
df_crop['Crop'] = df_crop['Crop'].replace('cotton(lint)','cotton')
df_crop['Crop'] = df_crop['Crop'].replace('moong(green gram)','mungbean')
df_crop['Crop'] = df_crop['Crop'].replace('urad','blackgram')
df_crop['Crop'] = df_crop['Crop'].replace('arhar/tur','pigeonpeas')
df_crop['Crop'] = df_crop['Crop'].replace('masoor','lentil')
df_crop['Crop'] = df_crop['Crop'].replace('gram','chickpea')
df_crop['Crop'] = df_crop['Crop'].replace('moth','mothbeans')
df['label'] = df['label'].replace('millet','small millets')
```

```python
labels = df['label'].unique()
crops = df_crop['Crop'].unique()
crop_list = []
for crop in labels:
    if crop in crops:
        crop_list.append(crop)
print(crop_list)
```

```python
df_crop = df_crop.drop(df_crop[~df_crop['Crop'].isin(crop_list)].index)
df_crop = df_crop.reset_index()
df = df.drop(df[~df['label'].isin(crop_list)].index).reset_index()
df = df.drop('index',axis = 1)
```

```python
rows = []
for crop in crop_list:
    t = df_crop[df_crop['Crop']==crop].sort_values(by='Yield',ascending=False)
    #print(t)
    t = t.iloc[0:t.shape[0]//3]
    so2 = t['SO2'].mean()
    no2 = t['NO2'].mean()
    rspm = t['RSPM'].mean()
    spm = t['SPM'].mean()
    rows.append([crop, so2, no2, rspm, spm])
temp = pd.DataFrame(rows, columns = ['Crop', 'SO2', 'NO2', 'RSPM', 'SPM'])
```

```python
temp
```

```python
import random
final_val = pd.DataFrame(columns=['SO2','NO2','RSPM', 'SPM'])
for i in range(0,df.shape[0]):
    crop = df.label.iloc[i]
    so2 = temp[temp['Crop']==crop]["SO2"].iloc[0]+random.randint(-3,3)
    no2 = temp[temp['Crop']==crop]["NO2"].iloc[0]+random.randint(-3,3)
    rspm = temp[temp['Crop']==crop]["RSPM"].iloc[0]+random.randint(-20,20)
    spm = temp[temp['Crop']==crop]["SPM"].iloc[0]+random.randint(-20,20)
    temp_dict = {'SO2':so2,'NO2':no2,'RSPM':rspm,'SPM':spm}
    
    final_val = final_val.append(temp_dict,ignore_index=True)
```

```python
final_val
```

```python
df["SO2"] = final_val["SO2"]
df["NO2"] = final_val["NO2"]
df["RSPM"] = final_val["RSPM"]
df["SPM"] = final_val["SPM"]
```
